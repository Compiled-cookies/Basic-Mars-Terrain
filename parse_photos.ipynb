```python
from PIL import Image
import requests

path_prefix = "https://mw1.google.com/mw-planetary/mars/elevation/"

example_block = "tttsrt"

image = Image.open(requests.get(f"{path_prefix}{example_block}.jpg", stream=True).raw)
width, height = image.size

for x in range(256):
    for y in range(256):
        curr = image.getpixel((x, y))
        
        minn, best_index = 1e9 + 7, 0
        
        for i, color_tuple in enumerate(convert_to_height):
            dist = (color_tuple[0] - curr[0])**2 + (color_tuple[1] - curr[1])**2 + (color_tuple[2] - curr[2])**2
            if dist < minn:
                minn = dist
                best_index = i
            
        print(best_index)
```
